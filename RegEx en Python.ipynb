# Expresiones Regulares en Python con RegEx

## Sintaxis de Expreciones regulares en Re

Las expresiones regulares pueden contener tanto caracteres especiales como ordinarios. La mayoría de los caracteres ordinarios, como 'A', 'a', o '0' son las expresiones regulares más sencillas; simplemente se ajustan a sí mismas. Se pueden concatenar caracteres ordinarios, así que last coincide con la cadena 'last'. (En el resto de esta sección, se escribirán los RE en , normalmente sin comillas, y las cadenas que deban coincidir .)este estilo especial'entre comillas simples'

Algunos caracteres, como `'|'` o `'('`, son especiales. Los caracteres especiales representan clases de caracteres ordinarios, o afectan a la forma en que se interpretan las expresiones regulares que los rodean.

Los operadores de repetición o cuantificadores `(*, +, ?, {m,n}, etc.)` no pueden ser anidados directamente. Esto evita la ambigüedad con el sufijo modificador no codicioso ?, y con otros modificadores en otras implementaciones. Para aplicar una segunda repetición a una repetición interna, se pueden usar paréntesis. Por ejemplo, la expresión `(?:a{6})` coincide con cualquier múltiplo de seis caracteres 'a'.

## Estructura de re.xxx

\- Toda función RegEx recibe por parámetro 3 partes: `re.findall("Patrón", "Texto", Flags)`
- Patron es la expecion regular que se dea buscar: `r'[a-z]+'`
- El texto a analizar.
- Flags son modificadoes de la expreciones ejemplo: `r.MULTILINE`

## Funciones RegEx:

- `re.match()` verifica una coincidencia solo al principio de la cadena.

- `re.search()` busca la primera coincidencia del patrón en cualquier parte de la cadena.

- `re.findall()` devuelve una lista con todas las coincidencias no superpuestas del patrón.

- `re.finditer()` devuelve un iterador con objetos de coincidencia para todas las ocurrencias en la cadena.

- `re.fullmatch()` verifica si la cadena completa coincide exactamente con el patrón de principio a fin

- `re.split()` divide la cadena en cada coincidencia del patrón y devuelve una lista con los fragmentos

- `re.sub()` reemplaza las coincidencias encontradas por una cadena de texto especificada `(r'\d', '*', texto)`

- `re.subn()` realiza los mismos reemplazos que re.sub(), pero devuelve una tupla con la nueva cadena y el número de sustituciones hechas

- `re.compile()` compila un patrón en un objeto de expresión regular para reutilizarlo de forma más eficiente

- `re.escape()` escapa todos los caracteres especiales de una cadena para que se interpreten literalmente en el patrón

- `re.purge()` limpia la memoria caché interna de las expresiones regulares compiladas recientemente

## Flags:

- `re.MULTILINE()`: Se usa para poder trabajar con las difiernte lineas ignorando los saltos de linea.
- `re.IGNORECASE()`: Ignora las mayusculas y minusculas.

\- Para poder usar mas de 1 flags se usa `"|"` para concatenar mas flags. <br> `"re.search(r'ExprecionRegular', texto, flahs = re.MULTILINE | re.IGNORECASE)`

## Objetos de patrón:

### Pattern Objects - generados por `re.compile()`):

- `.match()`, `.search()`, `.findall()`, etc.: realizan las mismas acciones que sus versiones del módulo re, pero se aplican directamente usando el patrón ya compilado, lo que mejora el rendimiento si se usan repetidamente.

- `.pattern()`: devuelve la cadena de texto original que se utilizó para compilar la expresión regular.

- `.flags()`: devuelve el valor numérico (entero) que representa los modificadores o flags aplicados a este patrón.

- `.groups()`: devuelve la cantidad de grupos de captura (definidos con paréntesis ()) que existen dentro del patrón.

- `.groupindex()`: devuelve un diccionario que mapea los nombres de los grupos de captura (si usaste grupos con nombre) a sus números correspondientes.

#### Funciones de los Pattern Objects

### Operadores regulares ? xd (Cuantificador en conjuntos)

- `"[]"` son para aceptar cualquier tipo de palabra con los caracter dentro de corchetes.
- `"+"` el patron tinee que coincidir una o mas veces `"[a-z]+"` o `"\d+"`.
- `"*"` el patron tiene que coincidir 0 a mas veces.
- `"?"` el patron tiene que coincidir 0 o 1 vez.
- `"{}"` se usa para especificar la cantidad de vez que quieres espeficicamente.
- `"{x,y}"` se usar para mostrar rangos.
- `"\b"` busca final o incio de palabra.
-  `"[^]"` es para negar.

\- Extra:
- `"(expresionRegular)" ` Esto srive para crear grupos.
-  `"^"` es para indicar el incio de una palabra.
-  `"$"` es para indicar al final de una palabra.
- `"""` la triple comilla es para poner texto plano hasta volver a cerra el triple comilla dobles.

### Algunas palabras reservadas

- `"\d"` es pra considera rangos del 0 al 9
    - Su inverso es `"\D"` donde regresa todo lo que no sean numeros
- `"w"` regresado todo lo que sea palabras de un idioma
    - `"W"` esto regresa su invesro simbolos, operadoes, saltos de linea, etc
- `"s"` consideras espacios de unicode, como espacios, saltos de linea, tabulador, retorno de carro
    -  `"S"` retornoa todo lo que no sea espacios

```python
import re

texto = "Hola\nMundo"

# 1. Por defecto (ignora el \n)
resultado_1 = re.findall(r".", texto)
print("Por defecto: ", resultado_1)
# Imprime: ['H', 'o', 'l', 'a', 'M', 'u', 'n', 'd', 'o']
# Nota: Falta el salto de línea.

# 2. Con la bandera DOTALL (incluye el \n)
resultado_2 = re.findall(r".", texto, flags=re.DOTALL)
print("Con DOTALL:  ", resultado_2)
# Imprime: ['H', 'o', 'l', 'a', '\n', 'M', 'u', 'n', 'd', 'o']
# Nota: El '\n' ahora sí está en la lista.

# 3. Con el modificador en línea (?s:.)
resultado_3 = re.findall(r"(?s:.)", texto)
print("En línea:    ", resultado_3)
# Imprime: ['H', 'o', 'l', 'a', '\n', 'M', 'u', 'n', 'd', 'o']
```
